### üßπ 2. Data Cleaning & Pre-processing

### üìà 4. Visualizations

### üìò univariant Analysis
‚Ä¢ Univariate analysis focuses on one variable at a time.                           
‚Ä¢ It helps us understand the distribution, spread, central tendency, and frequency of
individual features.

In this project, the following univariate plots were used:

```python
# üìâ 1.Histogram ‚Äì Price Distribution
sns.histplot(df['price'], bins=30)
plt.title("Price Distribution of Mobile Phones")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

### üìùInsights
‚óè Most phones fall within the lower to mid-price range, showing that the market is dominated by affordable and mid-range smartphones.

‚óè Very high-priced models appear much less frequently, indicating that premium flagship phones represent only a small portion of the dataset.

‚óè The distribution is right-skewed, meaning a few very expensive phones increase the spread, while the majority remain within an affordable price band.

```python
# üì¶ 2.Box Plot ‚Äì Price Outliers
sns.boxplot(data=df, x=df['price'])

plt.title("Price Outlier Analysis")
plt.xlabel("Price")
plt.show()
```

### üìùInsights
‚óè The boxplot shows several high-price outliers, indicating that a few premium flagship models are priced significantly above the general market range.

‚óè Most phones fall within a mid-range price band, suggesting that the majority of devices target budget-friendly and mid-budget consumers.

‚óè The long upper whisker compared to the lower whisker shows the price distribution is right-skewed, meaning expensive phones are fewer but extend far beyond the median price.

```python
# üîµ 3.Countplot ‚Äì Top 10 Mobile Count by Operating System
top_os = df['os'].value_counts().head(10)

sns.countplot(x=top_os.index)
plt.title("Top 10 Mobile Operating Systems by Count")
plt.xticks(rotation=65)
plt.show()
```

### üìùInsights
‚óè One operating system clearly dominates the market, showing that most smartphones in the dataset run on a single OS (likely Android).

‚óè Other operating systems have significantly fewer models, indicating limited availability or lower adoption in the market.

‚óè The skewed distribution highlights consumer and manufacturer preference, where the majority of brands focus on producing phones with the most popular OS.

```python
#üîµ 4. Histogram ‚Äì Battery Capacity Distribution
plt.hist(df['battery_clean'], bins=20)
plt.title("Battery Capacity Distribution")
plt.xlabel("Battery (mAh)")
plt.ylabel("Count")
plt.show()
```

### üåê Multivariate Analysis
‚Ä¢ Multivariate analysis examines three or more variables at the same time.             
‚Ä¢ This helps detect complex patterns, hidden relationships, and combined effects.

In this project, the following Multivariate plots were used:

```python
# üî• 13. Heatmap ‚Äì Correlation Matrix
numeric_cols = ['price', 'spec_score', 'user_rating', 'ram_clean', 'battery_clean']
numeric_df = df[numeric_cols]
corr = numeric_df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"size":8}, linewidths=0.5)
plt.title("Correlation Matrix of Mobile Phone Features")
plt.show()
```

### üìù Insights
‚óè Spec score shows a strong positive correlation with price, meaning higher-spec phones tend to be priced higher.

‚óè RAM and spec score are moderately/highly correlated, indicating RAM plays a key role in overall performance scoring.

‚óè Battery capacity has a weak or minimal correlation with price, showing brands don‚Äôt increase price solely based on battery size.

‚óè User rating has a low correlation with other technical specs, suggesting customer satisfaction depends on more than just raw specs.

‚óè RAM and battery have very low correlation, meaning phones with big batteries don‚Äôt necessarily come with higher RAM.

```python
# 14. üî∑ Pairplot
numeric_cols = ['price', 'spec_score', 'user_rating', 'ram_clean', 'battery_clean', 'display_size']

sns.pairplot(df[numeric_cols])
plt.show()
```

### üìù Insights
‚óè Price shows strong upward patterns with spec score and RAM, indicating that higher-spec devices are consistently priced higher.

‚óè Battery capacity has no clear visual relationship with price, showing battery size doesn‚Äôt directly influence pricing across brands.

‚óè Spec score and RAM display a positive trend, meaning devices with higher RAM tend to achieve better overall performance scores.

‚óè User rating does not show a strong trend with technical specs, suggesting ratings depend more on user experience factors than raw hardware.

‚óè Display size shows mild relationships with a few features, but no strong linear trend is visible, indicating screen size varies across all price and spec segments.

### ‚óâ 3D Scatter Plot

This 3D scatter plot visualizes how price, RAM, and battery capacity vary together across smartphones. It highlights how different models are positioned across budget, mid-range, and premium segments based on their technical features.

```python
# 3D ‚óèScatter Plot (Price vs RAM vs Battery)

fig = px.scatter_3d(
    df,
    x='price',
    y='ram_clean',
    z='battery_clean',
    opacity=0.7,
    title='3D Scatter Plot: Price vs RAM vs Battery Capacity'
)

fig.update_layout(
    width=1000,
    height=700,
    margin=dict(l=60, r=60, t=80, b=120),
    scene=dict(
        xaxis=dict(title='Price'),
        yaxis=dict(title='RAM (GB)'),
        zaxis=dict(title='Battery Capacity (mAh)')
    )
)

fig.show()
```

### üìù Insights
‚óè There is a positive relationship between price, RAM, and battery capacity, indicating 
that higher-priced phones typically offer better specifications.

‚óè Phones with high spec scores form a noticeable cluster, showing strong performance 
with balanced RAM and battery levels.

‚óè Budget phones create a dense group at low RAM and battery, reflecting limited 
specifications in low price ranges.

‚óè Mid-range phones often show good user ratings, even without high prices, 
highlighting strong value-for-money devices.

‚óè The scatter plot clearly separates budget, mid-range, and flagship phones based on 
their combined specification levels.

###  üìà 3D Line Plot
This 3D line plot shows the overall trend between price, RAM, and specification 
score. It helps illustrate how performance steadily increases as smartphones move
from budget to premium segments.

```python
# 3D üìàLine Trend: Price vs RAM vs Spec Score

fig = px.line_3d(
    df,
    x='price',
    y='ram_clean',
    z='spec_score',
    title='3D Line Trend of Mobile Specs'
)


fig.update_layout(
    width=1000,
    height=700,
    margin=dict(l=60, r=60, t=80, b=120),
    scene=dict(
        xaxis_title='Price',
        yaxis_title='RAM (GB)',
        zaxis_title='Spec Score'
    )
)

fig.show()
```

### üìù Insights
‚óè The line shows a consistent upward pattern, meaning that higher-priced phones
generally offer more RAM and improved specifications.

‚óè The relationship between RAM and spec score is positive, indicating that performance
increases with memory capacity.

‚óè Budget phones show a gradual increase in spec score, while mid-range phones show 
a stronger upward trend.

‚óè Flagship phones cause a sharp rise in the line, demonstrating a significant jump in 
performance at higher price levels.

‚óè Overall, the trend reveals that all three features ‚Äî price, RAM, and spec score ‚Äî 
grow proportionally, confirming a strong multivariate relationship.

## 5. Insight Generation and Report

### üìå Key Insights from the Analysis

### 1. Price Distribution & Market Positioning

‚óè The Indian smartphone market in 2025 is largely dominated by budget and mid-range devices, with most models priced below ‚Çπ20,000.

‚óè A smaller group of flagship smartphones form the top price tier, indicating a premium niche controlled by select brands.

‚óè A few price outliers represent special-edition, gaming-focused, or ultra-premium devices.

### 2. Specification Score Trends

‚óè There is a strong positive relationship between price and spec score, meaning higher performance devices generally cost more.

‚óè However, several value-for-money models offer powerful specifications at mid-range prices, reflecting aggressive pricing strategies by newer or competitive brands.

### 3. User Rating Patterns

‚óè Most smartphones receive user ratings between 3.5 and 4.5, indicating overall high customer satisfaction across segments.

‚óè Premium or high-priced phones do not always receive the best ratings, showing that brand value alone doesn‚Äôt guarantee user satisfaction.

‚óè Ratings are more influenced by performance (processor), camera quality, and brand trust, rather than battery size or display alone.

### 4. Brand-wise Performance

‚óè Some brands show a consistent price-to-performance ratio, maintaining steady quality across models.

‚óè Brands with a wide price range typically offer multiple series (budget ‚Üí mid-range ‚Üí flagship), reflecting strong market segmentation strategies.

### 5. Feature Relationships (Multivariate Insights)

 ‚óè Spec score rises with price, confirming clear segmentation by performance tiers.

 ‚óè Battery capacity shows a weak relationship with RAM, display size, or user ratings‚Äîindicating it does not heavily influence buying decisions.

 ‚óè Display size has minimal impact on price or spec score, making it more of a preference-based feature.

 ‚óè RAM and processor speed contribute significantly to spec score, reaffirming their importance in evaluating real-world smartphone performance.

### üéØ Significant Patterns and Anomalies

‚óè Some phones achieve high user ratings despite having only mid-range specification scores, indicating that customers often prioritize real-world performance, software optimization, stability, and overall user experience over raw hardware power.

‚óè A few models are priced significantly higher than their specification levels, suggesting the influence of brand value, marketing premium, or the continued pricing of older flagship models. These devices deviate from the typical price‚Äìperformance relationship.

‚óè Certain smartphones offer exceptionally high specification scores at relatively low prices, showing aggressive competitive pricing from newer or fast-growing brands. These devices serve as positive outliers and offer strong value for money.
